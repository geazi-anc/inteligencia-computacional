```python
import pandas as pd
df = pd.read_csv('AUS_albury.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df_removed = df.dropna(subset=['MinTemp','Rainfall','MaxTemp',
                               'WindGustSpeed','WindGustDir'])
```

```python
from sklearn.preprocessing import StandardScaler

#X = df.drop(columns=['RISK_MM', 'RainTomorrow'])
X = df_removed[['MinTemp','Rainfall','MaxTemp',
                'WindGustSpeed','WindGustDir']]

X['MeanTemp'] = (X['MinTemp'] + X['MaxTemp']) / 2
X = pd.get_dummies(X, columns=['WindGustDir'])

# normalizacao
scaler = StandardScaler()
scaler.fit(X)
X = scaler.transform(X)

y = df_removed.RainTomorrow
```

```python
X.columns
```

```python
X.isna().sum()
```

```python
X.shape, y.shape
```

```python
limite = int(len(X) * 0.7)
X_train, X_test = X[:limite], X[limite:]
y_train, y_test = y[:limite], y[limite:]
```

```python
from sklearn.neighbors import KNeighborsClassifier

modelo = KNeighborsClassifier(n_neighbors = 3)
modelo.fit(X_train, y_train)
```

```python
y_pred = modelo.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_pred, y_test)
```
